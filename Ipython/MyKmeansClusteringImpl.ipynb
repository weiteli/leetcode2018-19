```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("white")
import numpy as np
import random
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data[:, :2]
y = iris.target

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1,1,1)
colors = ['r','g', 'b']
label = [0, 1, 2]
for i in range(len(y)):
    ax.scatter(X[i][0], X[i][1], c=colors[y[i]], s=50)
ax.legend(label)
ax.axis('equal')
ax.grid()
```

```python
class MyKmeans():
    def __init__(self, k, max_iter):
        self.k = k
        self.max_iter = max_iter
    
    def _initializeCentroid(self, X):
        n_sample, n_dim = X.shape
        centroids = np.zeros((self.k, n_dim))
        
        for i in range(self.k):
            c = X[np.random.choice(range(n_sample)), :]
            centroids[i] = c

        return centroids
    
    def getClosestCentroid(self, x, centroids):
        dist = float('inf')
        label = 0
        for i, c in enumerate(centroids):
            cur_dist = np.linalg.norm(x-c)
            if  cur_dist < dist:
                label = i
                dist = cur_dist
        return label                
            
    def calculateNewCentroid(self, cluster):
        new_centroid = []
        for i in range(self.k):
            n = np.mean(X[cluster[i]], axis = 0)
            new_centroid.append(n)
        return new_centroid
    
    def getClusterLabel(self, clusters, X):
        y_pred = np.zeros(np.shape(X)[0])
        for cluster_i, cluster in enumerate(clusters):
            for sample_i in cluster:
                y_pred[sample_i] = cluster_i
        return y_pred
    
    def run(self, X):
        centroids = self._initializeCentroid(X)
        clusters = [[] for i in range(self.k)]

        for i in range(self.max_iter):
            for idx, x in enumerate(X):
                c = self.getClosestCentroid(x, centroids)
                clusters[c].append(idx)
            
            prev = centroids
            # calculate new centroid
            centroids = self.calculateNewCentroid(clusters)
            # Calculate centroid shift
            shift = np.array(centroids) - np.array(prev)
            err = np.linalg.norm(shift)
            if shift.any() == 0:
                break
        return centroids, self.getClusterLabel(clusters, X)
        
```

```python
my_kmeans = MyKmeans(2, 100)
cen, clusters = my_kmeans.run(X)
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1,1,1)
colors = ['r','g', 'b']
label = [0, 1, 2]
clusters = list(map(int, clusters.tolist()))

for i in range(len(clusters)):
    ax.scatter(X[i][0], X[i][1], c=colors[clusters[i]], s=50)

for i in range(len(cen)):
    ax.scatter(cen[i][0], cen[i][1], c=colors[i], marker='^', s=100)
ax.legend(label)
ax.axis('equal')
ax.grid()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1,1,1)
colors = ['r','g', 'b']
label = [0, 1, 2]
ax.scatter(X[:, 0], X[:, 1],c=kmeans.labels_, cmap='rainbow', s=50)
ax.legend(label)
ax.axis('equal')
ax.grid()
```
