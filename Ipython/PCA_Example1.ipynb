```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("white")
import pandas as pd
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

iris = datasets.load_iris()
X = iris.data[:, :4]
y = iris.target
X_norm = StandardScaler().fit_transform(X)
# Currently, we have data for 4 dimensions
# We will try to project it to 2 dimensions
pca = PCA(n_components=2)
pcs = pca.fit_transform(X_norm)

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1,1,1)
ax.set_xlabel('Component 1')
ax.set_ylabel('Component 2')
colors = ['r','g','b']
label = [0, 1, 2]
print(pca.explained_variance_ratio_)
for i in range(len(y)):
    ax.scatter(pcs[i][0], pcs[i][1], c=colors[y[i]], s=50)
ax.legend(label)
ax.grid()
```

```python
from sklearn.datasets import fetch_mldata
from sklearn.cross_validation import train_test_split
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("MNIST_data/")
train_img = mnist.train.images
train_label = mnist.train.labels
test_img = mnist.test.images
test_label = mnist.test.labels
print(len(train_img))
print("READY")
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(train_img)
train_img = scaler.transform(train_img)
test_img = scaler.transform(test_img)
mnist_pca = PCA(0.90)
mnist_pca.fit(train_img)
print(mnist_pca.n_components_)
projected_train_img = mnist_pca.transform(train_img)
projected_test_img = mnist_pca.transform(test_img)
```

```python
index = 130
approcimation = mnist_pca.inverse_transform(projected_train_img[index])
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(1,2,1)
ax1.imshow(approcimation.reshape(28,28), cmap='jet', interpolation='nearest')
ax1 = fig.add_subplot(1,2,2)
ax1.imshow(train_img[index].reshape(28,28), cmap='jet', interpolation='nearest')
plt.show()
```

```python
import numpy as np
mnist_pca2 = PCA()
mnist_pca2.fit(train_img)
total = sum(mnist_pca2.explained_variance_ratio_)
var_curve = [mnist_pca2.explained_variance_ratio_[i]/total for i in range(mnist_pca2.n_components_)]
cum_var = np.cumsum(var_curve)
plt.figure(figsize=(10,5))
plt.step(range(1,mnist_pca2.n_components_+1), cum_var)
plt.ylabel('Cumulative Explained variance')
plt.xlabel('Principal components')
plt.axhline(y = 0.95, color='k', linestyle='--', label = '95% Explained Variance')
plt.axhline(y = 0.90, color='c', linestyle='--', label = '90% Explained Variance')
plt.axhline(y = 0.85, color='r', linestyle='--', label = '85% Explained Variance')
plt.legend(loc='best')
plt.show()
```
