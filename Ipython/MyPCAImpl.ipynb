```python
# Reference
# 1. https://sebastianraschka.com/Articles/2014_pca_step_by_step.html 
# 2. https://plot.ly/ipython-notebooks/principal-component-analysis/
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("white")
import pandas as pd
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

iris = datasets.load_iris()
X = iris.data[:, :4]
y = iris.target
X_norm = StandardScaler().fit_transform(X)
# Currently, we have data for 4 dimensions
# We will try to project it to 2 dimensions
pca = PCA(n_components=2)
pcs = pca.fit_transform(X)

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1,1,1)
ax.set_xlabel('Component 1')
ax.set_ylabel('Component 2')
colors = ['r','g','b']
label = [0, 1, 2]
print(pca.explained_variance_ratio_)
for i in range(len(y)):
    ax.scatter(pcs[i][0], pcs[i][1], c=colors[y[i]], s=50)
ax.legend(label)
ax.grid()
```

```python
import numpy as np
mnist_pca2 = PCA()
mnist_pca2.fit(train_img)
total = sum(mnist_pca2.explained_variance_ratio_)
var_curve = [mnist_pca2.explained_variance_ratio_[i]/total for i in range(mnist_pca2.n_components_)]
cum_var = np.cumsum(var_curve)
plt.figure(figsize=(10,5))
plt.step(range(1,mnist_pca2.n_components_+1), cum_var)
plt.ylabel('Cumulative Explained variance')
plt.xlabel('Principal components')
plt.axhline(y = 0.95, color='k', linestyle='--', label = '95% Explained Variance')
plt.axhline(y = 0.90, color='c', linestyle='--', label = '90% Explained Variance')
plt.axhline(y = 0.85, color='r', linestyle='--', label = '85% Explained Variance')
plt.legend(loc='best')
plt.show()
```

```python
class MyPCA():
    def __init__(self, n_components):
        self.n_components = n_components
        
    def fit(self, X):
        self.X_mean = np.mean(X, axis=0)
        #self.X_std = np.std(X, axis=0)
        X = (X - self.X_mean)      
        self.cov = X.T.dot(X) / (X.shape[0])
        self.eig_vals, self.eig_vecs = np.linalg.eig(self.cov)
        print("Before Transpose\n", self.eig_vecs)
        self.eig_vecs = self.eig_vecs.T
        print("After Transpose\n", self.eig_vecs)
        for ev in self.eig_vecs:
            np.testing.assert_array_almost_equal(1.0, np.linalg.norm(ev))
        
        self.eigen_pairs = [(np.abs(self.eig_vals[i]), self.eig_vecs[i]) for i in range(len(self.eig_vals))]
        self.eigen_pairs.sort(key=lambda x: x[0])
        self.eigen_pairs.reverse()
        print(self.eigen_pairs[0][1].dot(self.eigen_pairs[1][1]))
        
    def transform(self, X):
        self.projection_matrix = []
        dim = len(X[0])
        for i in range(self.n_components):
            print(self.eigen_pairs[i][0])
            print(self.eigen_pairs[i][1])
                
            if len(self.projection_matrix) == 0:
                self.projection_matrix = self.eigen_pairs[i][1].reshape(dim,1)
            else:
                self.projection_matrix = np.hstack((self.projection_matrix, self.eigen_pairs[i][1].reshape(dim,1)))
        return (X-self.X_mean).dot(self.projection_matrix)
        
        
```

```python
mpca = MyPCA(2)
mpca.fit(X)
projected_X2D = mpca.transform(X)
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1,1,1)
for i in range(len(y)):
    ax.scatter(projected_X2D[i][0], projected_X2D[i][1], c=colors[y[i]], s=50)
plt.show()
```
