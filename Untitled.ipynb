```python
image_gray = cv2.cvtColor(opencv_image, cv2.COLOR_BGR2GRAY)
ret, thres = cv2.threshold(image_gray, 127, 255, 0)
image, contours, hier = cv2.findContours(thres, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
counts = sorted([(c, cv2.boundingRect(c)[0]) for c in contours], key=lambda x:x[1])
ary = []
for (c,_) in counts:
    (x,y,w,h) = cv2.boundingRect(c)
    print((x,y,w,h))
    if w >= 15 and h == 24:
        ary.append((x,y,w,h))
```
